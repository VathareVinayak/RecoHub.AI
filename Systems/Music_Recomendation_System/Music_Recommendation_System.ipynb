```python
data['Singer/Artists'] = data['Singer/Artists'].str.replace(' , ',' ')
```

```python
dataset = data[["Song-Name" , "tags"]]
dataset
```

```python
dataset.head()
```

```python
dataset['Song-Name'].value_counts().head(10)
```

```python
dataset["tags"] = dataset["tags"].astype(str).apply(lambda x: x.lower())
dataset
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=2000)
vectors = cv.fit_transform(dataset["tags"]).toarray()
```

```python
cv
```

```python
vectors.shape
```

```python
# Replace cv.get_feature_names() with cv.get_feature_names_out()
feature_names =  cv.get_feature_names_out()
feature_names
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
sorted(list(enumerate(similarity[0])),reverse=True , key = lambda x:x[1])
```

```python
dataset.rename(columns={"Song-Name": "title"}, inplace=True)
```

```python
# Function to get recommendations
def recommend_music(music):
    music_index = dataset[dataset["title"] == music].index[0]
    distances = similarity[music_index]
    music_list = sorted(list(enumerate(distances)),reverse=True , key = lambda x:x[1])[1:11]
    for i in music_list:
        print(dataset.iloc[i[0]]["title"])
    
```

```python
recommend_music('Aankh Marey')
```
