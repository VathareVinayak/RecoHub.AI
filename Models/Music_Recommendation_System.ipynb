The song with the **highest** user **Rating**

```python
max_rating = data['User-Rating'].max()
song_with_max_rating = data[data['User-Rating'] == max_rating]['Song-Name'].iloc[0]

print(f"The song with the highest user rating is: {song_with_max_rating}")
```

** THE SONG WITH LOWEST USER RATING **

```python
max_rating = data['User-Rating'].min()
song_with_max_rating = data[data['User-Rating'] == max_rating]['Song-Name'].iloc[0]

print(f"The song with the lowest user rating is: {song_with_max_rating}")
```

```python
dataset["tags"] = dataset["tags"].astype(str).apply(lambda x: x.lower())
dataset
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```
